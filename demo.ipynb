```python
def format_result2(df):
    datasets = ['AID',"DOTA","DIOR"]
    metrics = ['FID',"LPIPS"]
    methods = ['Bicubic','LIIF','SADN',"IDM","E2DiffSR"]
    def get_val(dataset, method, scale, metric):
        for idx, row in df.iterrows():
            if row['methodName'].lower() == method.lower() and row['dataset'] == dataset and row['degradName'] == scale:
                return row[metric.lower()]
    scales = [2.0,2.3,2.6,3.0,3.4,4.0,6.0,8.0,10.0,12.0]
    
    results = []
    for dataset in datasets:
        for method in methods:
            for metric in metrics:
                rslt = {
                        'Datasset': dataset,
                        'Methods': method,
                        'Metric': metric,
                    }
                for s in scales:
                    rslt[f'x{s:.1f}'] = get_val(dataset, method, f'x{s:.1f}', metric)
                results.append(rslt)

    formatted_df = pd.DataFrame(results)
    return formatted_df

format_result2(df)
```
